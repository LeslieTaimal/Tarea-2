```python
import re
import csv
```

```python
with open("pubmed-crisprcasn-set.txt", errors="ignore") as l: 
    my_text = l.read()
```

```python
len(my_text)
```

```python
my_text = re.sub(r'\n\s{6}', ' ', my_text)
```

```python
print(my_text[:2000])
```

```python
 x = re.split('^AD+[\D]+[.=$]', my_text)
```

```python
print(x)
```

```python
country = re.findall(r'[,]+[\s]+[\w]+[.=$]|[,]+[\s]+[\w]+[\s]+[\w]+[.=$]', my_text)
```

```python
len(country)
```

```python
unique_country = list(set(country))
```

```python
unique_country
```

```python
paises=unique_country
```
